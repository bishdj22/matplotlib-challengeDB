```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data = pd.read_csv("../data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("../data/clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data(*SEE ABOVE*)


# Combine the data into a single dataset
merge_mousedata = pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID", how="outer")
merge_mousedata.head()

# Display the data table for preview(*SEE ABOVE*)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
meantumor_data = merge_mousedata.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)']
avg_tumorvol = meantumor_data.mean()

# Convert to DataFrame
avgtumor_df = avg_tumorvol.reset_index()
# Preview DataFrame
avgtumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#sem_tumorvol = merge_mousedata.groupby(['Drug','Timepoint'])
std_errortumor = meantumor_data.sem()
#Drug_col = sem_tumorvol["Drug"]

# Convert to DataFrame
stderror_tumorvol_df = pd.DataFrame(std_errortumor)

# Preview DataFrame

stderror_tumorvol_df.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#GOOD CODE --> for pivoting, table issue is missing values in columns 

pivot_avgtumordf = avgtumor_df.pivot(index='Timepoint', columns='Drug',values="Tumor Volume (mm3)").reset_index()

# Preview that Reformatting worked
pivot_avgtumordf
```

```python
#Rearrange data frame to only show drugs to be analyzed by the error plot graph
plot_tumorpivot = pivot_avgtumordf.drop(["Naftisol","Ceftamin","Propriva","Ramicane","Stelasyn","Zoniferol"],axis=1)
plot_tumorpivot
```

```python
# Generate the Plot (with Error Bars)
#List of the drugs we'll be analyzing

drug_format_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]

#For loop to plot graph

for drug,marker,colors in drug_format_list:

    x_axis = plot_tumorpivot["Timepoint"]
    y_axis = plot_tumorpivot[drug]
    err = std_errortumor[drug]

    bs_plt = plt.errorbar(x_axis,y_axis,err,fmt=marker,ls='--',color=colors,linewidth=0.5,label=drug)
    
#Format graph, legend, axes   
plt.legend(loc='best')
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

#Show Figure
plt.grid()

# Save the Figure **** NEED TO DO THIS ***
plt.savefig("../PymaceuticalsDB/tumor_responsegraph.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_data = merge_mousedata.groupby(['Drug','Timepoint'])['Mouse ID']
micecount_grouped = survival_data.count()

# Convert to DataFrame
mice_group_df = micecount_grouped.reset_index()

# Convert to DataFrame

# Preview DataFrame
mice_group_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mice_groups = mice_group_df.pivot(index='Timepoint', columns='Drug',values="Mouse ID").reset_index()
# Preview that Reformatting worked
pivot_mice_groups.head()
# Preview the Data Frame
```

```python
plot_survivalpivot = pivot_mice_groups.drop(["Naftisol","Ceftamin","Propriva","Ramicane","Stelasyn","Zoniferol"],axis=1)
plot_survivalpivot
```

```python
# Generate the Plot (Accounting for percentages)
survival_format_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]

#For loop to plot graph

for drug,marker,colors in survival_format_list:

    x_axis = plot_survivalpivot["Timepoint"]
    y_axis = plot_survivalpivot[drug]/25
    err = std_errormeta[drug]

    bs_plt = plt.errorbar(x_axis,y_axis*100,err,fmt=marker,ls='--',color=colors,linewidth=0.5,label=drug)
    
#Format graph, legend, axes   
plt.legend(loc='best')
plt.title('Survival Rate During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')



# Show the Figure
plt.grid()

# Save the Figure
plt.savefig("../PymaceuticalsDB/survival_ratechart.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

percent_chg =((pivot_avgtumordf.loc[9] - pivot_avgtumordf.loc[0] )/ pivot_avgtumordf.loc[0])*100

percent_chg
```

```python
end_plot_pivot = avgtumor_df.pivot_table(index='Timepoint', columns='Drug',values="Tumor Volume (mm3)")
```

```python
percent_change = []
columns = list(end_plot_pivot.columns.values)

for drug in columns:
    percent_chg = ((end_plot_pivot.iloc[-1, end_plot_pivot.columns.get_loc(drug)] - end_plot_pivot.iloc
                    [0, end_plot_pivot.columns.get_loc(drug)]) / end_plot_pivot.iloc[0, end_plot_pivot.columns.get_loc(drug)]) * 100 
    percent_change.append(percent_chg)

percent_change_df = pd.DataFrame({"Drug" : columns,"Percent Change" : percent_change})
percent_change_df.drop(axis=1,index=0)
percent_change_df
```

```python
# Store all Relevant Percent Changes into a Tuple

capomulin_avg = percent_change_df.loc[percent_change_df.index[0], "Percent Change"]
infubinol_avg = percent_change_df.loc[percent_change_df.index[2], "Percent Change"]
ketapril_avg = percent_change_df.loc[percent_change_df.index[3], "Percent Change"]
placebo_avg = percent_change_df.loc[percent_change_df.index[5], "Percent Change"]

drug_list = ("Capomulin","Infubinol","Ketapril","Placebo")
avg_percent_chng = (capomulin_avg, infubinol_avg, ketapril_avg, placebo_avg)

success_df = pd.DataFrame(avg_percent_chng,drug_list, columns = ["Percent Change"])
success_df["passing"] = success_df["Percent Change"] < 0

#Plot Results

fig, ax = plt.subplots()
x_axis = np.arange(len(drug_list))

bars = ax.bar(x_axis, avg_percent_chng, color=success_df.passing.map({True: 'g', False:'r'}),width=1,align='center')

tick_locations = [x for x in x_axis]
ax.set_xticks(tick_locations)
ax.set_yticks(np.arange(-20,80,step=20))
ax.set_ylim(-30,70)
ax.set_ylabel('% Change in Tumor Size')
ax.set_title('Tumor Chg Over 45 Days Treatment')
ax.set_xticklabels(drug_list)
ax.grid()

# Use functions to label the percentages of changes

def autolabel(bars):
    
    for rect in ax.patches:
        y_value = rect.get_height() 
        x_value = rect.get_x() + rect.get_width() / 2
        label = "{:.2%}".format(y_value) 
        #'%d' % int(y_value)
        if y_value < 0:
            ax.text(x_value, y_value + 15, '%d' % int(y_value) + "%", ha='center', va='center', color='white',weight='bold')
        elif y_value > 55:
            ax.text(x_value, y_value - 56, '%d' % int(y_value) + "%", ha='center', va='bottom', color='white',weight='bold')
        elif y_value > 50:
            ax.text(x_value, y_value - 50, '%d' % int(y_value) + "%", ha='center', va='bottom', color='white',weight='bold')
        else:
            ax.text(x_value, y_value - 45, '%d' % int(y_value) + "%", ha='center', va='bottom', color='white',weight='bold')


# Call functions to implement the function calls

autolabel(bars)

# Save the Figure
plt.savefig("../PymaceuticalsDB/avg_tumor_chg.png")
```
